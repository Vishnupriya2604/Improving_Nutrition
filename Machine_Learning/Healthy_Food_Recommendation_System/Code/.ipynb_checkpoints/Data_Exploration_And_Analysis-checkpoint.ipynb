```python
#Load the necessary packages
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read all the data files 
branded_food = pd.read_csv(r'/Users/vishnupriyavenkateswaran/Desktop/GWU_Sem4/Code/Machine_Learning/Healthy_Food_Recommendation_System/Data/branded_food.csv')
food = pd.read_csv(r'/Users/vishnupriyavenkateswaran/Desktop/GWU_Sem4/Code/Machine_Learning/Healthy_Food_Recommendation_System/Data/food.csv')
food_nutrient = pd.read_csv(r'/Users/vishnupriyavenkateswaran/Desktop/GWU_Sem4/Code/Machine_Learning/Healthy_Food_Recommendation_System/Data/food_nutrient.csv')
food_attribute = pd.read_csv(r'/Users/vishnupriyavenkateswaran/Desktop/GWU_Sem4/Code/Machine_Learning/Healthy_Food_Recommendation_System/Data/food_attribute.csv')
nutrient = pd.read_csv(r'/Users/vishnupriyavenkateswaran/Desktop/GWU_Sem4/Code/Machine_Learning/Healthy_Food_Recommendation_System/Data/nutrient.csv')
```

```python
# Drop the id column
food_nutrient.drop(columns='id', inplace=True)
```

```python
#Rename the column
food_nutrient.rename(columns={'nutrient_id': 'id'}, inplace=True)
```

```python
#Data Cleaning

nutrient_combined = pd.merge(left=food_nutrient, right=nutrient, on='id')
nutrient_combined.head()
```

```python
#Positive needed: Protein, fiber, vitamin A, vitamin C, Calcium, Iron

#Negative needed: Saturated Fat, Added Sugar, Sodium

nutrient_combined.isnull().sum()

```

```python
# Drop the unwanted columns
nutrient_combined_2 = nutrient_combined.drop(columns=['data_points', 'derivation_id', 'min', 'max', 'median', 
                                                     'footnote', 'min_year_acquired', 'rank'], axis=1)
```

```python
nutrient_combined_2.head()
```

```python
nutrient_combined_2.isnull().sum()
```

```python
nutrient_combined_2.shape
```

```python
#Renaming the columns
protein = nutrient_combined_2[nutrient_combined_2['name']=='Protein']
fiber = nutrient_combined_2[nutrient_combined_2['name']=='Fiber, total dietary']
mufa = nutrient_combined_2[nutrient_combined_2['name']=='Fatty acids, total monounsaturated']
vitamin_a = nutrient_combined_2[nutrient_combined_2['name']=='Vitamin A, IU']
vitamin_c = nutrient_combined_2[nutrient_combined_2['name']=='Vitamin C, total ascorbic acid']
calcium = nutrient_combined_2[nutrient_combined_2['name']=='Calcium, Ca']
potassium = nutrient_combined_2[nutrient_combined_2['name']=='Potassium, K']
magnesium = nutrient_combined_2[nutrient_combined_2['name']=='Magnesium, Mg']
iron = nutrient_combined_2[nutrient_combined_2['name']=='Iron, Fe']
trans_fat = nutrient_combined_2[nutrient_combined_2['name']=='Fatty acids, total trans']
sat_fat = nutrient_combined_2[nutrient_combined_2['name']=='Fatty acids, total saturated']
sugar = nutrient_combined_2[nutrient_combined_2['name']=='Sugars, total including NLEA']
sodium = nutrient_combined_2[nutrient_combined_2['name']=='Sodium, Na']
```

```python
vitamin_c.head()
```

```python
#Reading the data file
branded_food.head()
```

```python
branded_combined = pd.merge(left=branded_food, right=food, on='fdc_id')
branded_combined.head()
```

```python
branded_combined.drop(columns=['household_serving_fulltext', 'food_category_id', 'gtin_upc', 
                               'data_source', 'modified_date', 'available_date'], inplace=True)
```

```python
branded_combined['cat_null'] = branded_combined['branded_food_category'].isnull()
```

```python
branded_combined.head()
```

```python
branded_combined.isnull().sum()
```

```python
#Rename columns

sat_fat.rename(columns={'amount': 'sat_fat_g'}, inplace=True)
sat_fat_g = sat_fat[['fdc_id', 'sat_fat_g']]
branded = pd.merge(branded_combined, sat_fat_g, on='fdc_id', how='outer')
```

```python
#Rename Columns
trans_fat.rename(columns={'amount': 'trans_fat_g'}, inplace=True)
trans_fat_g = trans_fat[['fdc_id', 'trans_fat_g']]
branded = pd.merge(branded, trans_fat_g, on='fdc_id', how='outer')
```

```python
magnesium.rename(columns={'amount': 'magnesium_mg'}, inplace=True)
magnesium_mg = magnesium[['fdc_id', 'magnesium_mg']]
branded = pd.merge(branded, magnesium_mg, on='fdc_id', how='outer')
```

```python
potassium.rename(columns={'amount': 'potassium_mg'}, inplace=True)
potassium_mg = potassium[['fdc_id', 'potassium_mg']]
branded = pd.merge(branded, potassium_mg, on='fdc_id', how='outer')
```

```python
mufa.rename(columns={'amount': 'mufa_grams'}, inplace=True)
mufa_grams = mufa[['fdc_id', 'mufa_grams']]
branded = pd.merge(branded, mufa_grams, on='fdc_id', how='outer')
```

```python
protein.rename(columns={'amount': 'protein_grams'}, inplace=True)
protein_grams = protein[['fdc_id', 'protein_grams']]
branded = pd.merge(branded, protein_grams, on='fdc_id', how='outer')
```

```python
fiber.rename(columns={'amount': 'fiber_grams'}, inplace=True)
fiber_grams = fiber[['fdc_id', 'fiber_grams']]
branded = pd.merge(branded, fiber_grams, on='fdc_id', how='outer')
```

```python
vitamin_a.rename(columns={'amount': 'vitamin_a_IU'}, inplace=True)
vitamin_a_IU = vitamin_a[['fdc_id', 'vitamin_a_IU']]
branded = pd.merge(branded, vitamin_a_IU, on='fdc_id', how='outer')
```

```python
vitamin_c.rename(columns={'amount': 'vitamin_c_MG'}, inplace=True)
vitamin_c_MG = vitamin_c[['fdc_id', 'vitamin_c_MG']]
branded = pd.merge(branded, vitamin_c_MG, on='fdc_id', how='outer')
```

```python
calcium.rename(columns={'amount': 'calcium_MG'}, inplace=True)
calcium_MG = calcium[['fdc_id', 'calcium_MG']]
branded = pd.merge(branded, calcium_MG, on='fdc_id', how='outer')
```

```python
iron.rename(columns={'amount': 'iron_MG'}, inplace=True)
iron_MG = iron[['fdc_id', 'iron_MG']]
branded = pd.merge(branded, iron_MG, on='fdc_id', how='outer')
```

```python
sugar.rename(columns={'amount': 'sugar_G'}, inplace=True)
sugar_G = sugar[['fdc_id', 'sugar_G']]
branded = pd.merge(branded, sugar_G, on='fdc_id', how='outer')
```

```python
sodium.rename(columns={'amount': 'sodium_MG'}, inplace=True)
sodium_MG = sodium[['fdc_id', 'sodium_MG']]
branded = pd.merge(branded, sodium_MG, on='fdc_id', how='outer')
```

```python
branded.head()
```

```python
branded.fillna(0, inplace=True)
```

```python
#Calculate nutritional index

branded['protein_grams'] = branded['protein_grams']/50
branded['fiber_grams'] = branded['fiber_grams']/25
branded['vitamin_a_IU'] = branded['vitamin_a_IU']/5000
branded['vitamin_c_MG'] = branded['vitamin_c_MG']/60
branded['calcium_MG'] = branded['calcium_MG']/1000
branded['iron_MG'] = branded['iron_MG']/18
branded['sugar_G'] = branded['sugar_G']/125
branded['sodium_MG'] = branded['sodium_MG']/2400
branded['mufa_grams'] = branded['mufa_grams']/20
branded['potassium_mg'] = branded['potassium_mg']/3500
branded['magnesium_mg'] = branded['magnesium_mg']/400
branded['trans_fat_g'] = branded['trans_fat_g']
branded['sat_fat_g'] = branded['sat_fat_g']/20
```

```python
#Algorithm used as per the mentioned research paper
branded['NRFNn.3'] = branded['protein_grams']+branded['fiber_grams']+branded['vitamin_a_IU']+branded['vitamin_c_MG']+branded['calcium_MG']+branded['mufa_grams']+branded['potassium_mg']+branded['magnesium_mg']+branded['iron_MG']-branded['trans_fat_g']-branded['sat_fat_g']-branded['sugar_G']-branded['sodium_MG']
```

```python
sns.pairplot(branded[['sat_fat_g', 'trans_fat_g', 'sugar_G', 'sodium_MG']])
```

```python
branded.protein_grams.clip(upper=1, inplace=True)
branded.fiber_grams.clip(upper=1, inplace=True)
branded.vitamin_a_IU.clip(upper=1, inplace=True)
branded.vitamin_c_MG.clip(upper=1, inplace=True)
branded.calcium_MG.clip(upper=1, inplace=True)
branded.iron_MG.clip(upper=1, inplace=True)
branded.potassium_mg.clip(upper=1, inplace=True)
branded.magnesium_mg.clip(upper=1, inplace=True)
branded.mufa_grams.clip(upper=1, inplace=True)
branded.sugar_G.clip(upper=1, inplace=True)
branded.sat_fat_g.clip(upper=1, inplace=True)
branded.sodium_MG.clip(upper=1, inplace=True)
branded.trans_fat_g.clip(upper=1, inplace=True)
```

```python
branded['NRFNn.3'] = branded['protein_grams']+branded['fiber_grams']+branded['vitamin_a_IU']+branded['vitamin_c_MG']+branded['calcium_MG']+branded['mufa_grams']+branded['potassium_mg']+branded['magnesium_mg']+branded['iron_MG']-branded['trans_fat_g']-branded['sat_fat_g']-branded['sugar_G']-branded['sodium_MG']
```
