```python
#Importing required python packages

import pandas as pd
from utils import read_and_clean_data, split_X_y
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Path to the Excel file.
path_to_data = "./data/Nutrients.xlsx"
dataset = pd.read_excel(path_to_data)
dataset.head()
```

```python
# Check variables 
dataset.columns
```

```python
# Check data type for each column
dataset.dtypes
```

```python
# Seperate Numerical and Categorical variable

tot_cols = dataset.columns
num_cols = dataset._get_numeric_data().columns
cat_cols = list(set(tot_cols)-set(num_cols))
cat_cols
```

```python
# Remove food indeces and names

drop_cols = ['NDB_No', 'Shrt_Desc']
dataset = dataset.drop(drop_cols, axis=1)
dataset
```

```python
dataset.dtypes
```

```python
#Drop unwanted columns
drop_cols = ['GmWt_Desc1', 'GmWt_Desc2']
dataset = dataset.drop(drop_cols, axis=1)
dataset
```

```python
dataset.columns
```

This is where we get the lables of our dataset. Accordign to the UNICEF, the main deficiencies in terms of micronutrients are Vitamin A, Iron, Zinc, Calcium, Vitamin D, and Folate

```python
# Define the labels
labels = ['Calcium_(mg)', 'Iron_(mg)','Zinc_(mg)', 'Vit_A_IU', 'Vit_D_IU', 'Folate_Tot_(µg)']
labels
```

```python
# Check for missing data

dataset.isnull().sum()
```

```python
# Fill missing data with median

get_median = dataset.median()
dataset = dataset.fillna(get_median)
```

```python
# Check for missing data again

dataset.isna().values.any()
```

```python
# Seperate the data into input and output data

# List of names of the columns that we will drop for the input data
drop_cols = ['Vit_D_µg', 'Vit_A_RAE'] + labels
# Get the data for labels
y = dataset[labels].values
# Get the data for inputs
X = dataset.drop(drop_cols, axis=1).values

```

```python
y.shape, X.shape
```

```python
#Correlation between variables

dataset.corr()
```

```python
#Correlation Plot
import seaborn as sns
sns.set(style="dark")
corr = dataset.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(20, 19))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5)
```
